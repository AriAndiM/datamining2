## 1. K-Means
>*K-Means clustering adalah algoritma unsupervised learning yang dipakai untuk mengelompokkan dataset yang belum dilabel ke dalam kluster yang berbeda. Simbol K pada K-means clustering menandakan jumlah kluster yang digunakan.*

>Langkah-langkah :

>1. Pilih angka K untuk menentukan jumlah cluster.
>2. Pilih titik K atau centroid secara acak.
>3. Tetapkan setiap titik data ke centroid terdekat, yang akan membentuk cluster K yang telah ditentukan.
>4. Hitung varians dan tempatkan centroid baru dari setiap cluster.
>5. Ulangi langkah ketiga, yang berarti menetapkan kembali setiap titik data ke centroid terdekat baru dari setiap cluster.
>6. Jika ada penugasan ulang, lanjutkan ke langkah-4 jika tidak, lanjutkan ke FINISH.
>7. Model sudah siap.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
nama_kolom = ['sepal-length','sepal-width','petal-length','petal-width','Class']
data_iris = pd.read_csv(url, names=nama_kolom)
```

```python
data_iris.head()
```

Converting nama class ke numerical 

```python
data_iris['Class'] = pd.Categorical(data_iris['Class'])
data_iris['Class'] = data_iris['Class'].cat.codes
```

Menyiapkan dataset

```python
x = data_iris.values[:, 0:4]
y = data_iris.values[:, 4]
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
y_true = le.fit_transform(y)
y_true
```

Menjalankan K-Means

```python
from sklearn.cluster  import KMeans

kmeans = KMeans(n_clusters=3)
kmeans = kmeans.fit(x)
labels = kmeans.predict(x)
centroid = kmeans.cluster_centers_
```

Mengecak akurasi clustering 

```python
from sklearn.metrics import classification_report
target_names = ['Iris-setosa','Iris-versicolor','Iris-virginica']
print(classification_report(data_iris['Class'],kmeans.labels_,target_names=target_names))
```
