2. Pisahkan data dan Latih Modelnya

Splitting data ke training dan test set (80:20)

```python
x=iris.data
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=4)
```

Shape train dan test objek

```python
print(X_train.shape)
print(X_test.shape)
```

```python
print(y_train.shape)
print(y_test.shape)
```

```python
#Import scikitlearnclassifier class from sklearn
from sklearn.neighbors import KNeighborsClassifier

#Import matriks model untuk mengetes akurasi
from sklearn import metrics

k_range = range(1,26)
scores={}
scores_list = []
for k in k_range:
  knn=KNeighborsClassifier(n_neighbors=k)
  knn.fit(X_train,y_train)
  y_pred=knn.predict(X_test)
  scores[k]=metrics.accuracy_score(y_test,y_pred)
  scores_list.append(metrics.accuracy_score(y_test,y_pred))
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

#plot hubungan antara K dan testing accuracy
plt.plot(k_range,scores_list)
plt.xlabel('Value Of K for KNN')
plt.ylabel('Testing Accuracy')
```
