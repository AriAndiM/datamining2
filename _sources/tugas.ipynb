### 1. Membaca data cuaca seattle dari url

```python
angin = data ['wind']
angin
```

### 2. Mencari mean,median,dan modus


```python
import statistics as st
```

2.1 Mean (nilai rata-rata)

>*Mean yaitu nilai rata-rata dari data*

>*   Rumus Mean :

$$Mean= \frac{x_{1}+x_{2}+x_{3}+...+x_{n}}{n}$$

>Mean = rata-rata

>$x_{1}$ = nilai data ke-1

>$x_{n}$ = nilai data ke-n

>$n_{}$ = banyaknya data


    



```python
st.mean(angin)
```

2.2 Median (nilai tengah)

>*Median yaitu nilai tengah dari data*

>Rumus jika jumlah data ganjil : 



>$$M_{e}=\frac{X(n+1)}{2}$$

>Rumus jika jumlah data genap : 



>$$M_{e}=\frac{X_{\frac{n}{2}}+X(_{\frac{n}{2}+1})}{2}$$






```python
st.median(angin)
```

2.3 Modus (Nilai paling sering muncul)

>*Modus yaitu nilai yang paling sering muncul pada data*



```python
st.mode(angin)
```

### 3. Kemiringan / Skewned

> *Kemencengan atau kecondongan (skewness) adalah tingkat ketidaksimetrisan atau kejauhan simetri dari sebuah distribusi. Skewness diartikan sebagai kemiringan distribusi data. Sebuah distribusi yang tidak simetris akan memiliki rata-rata, median, dan modus yang tidak sama besarnya sehingga distribusi akan terkonsentrasi pada salah satu sisi dan kurvanya akan menceng.*



```python
angin.skew()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure()
sns.distplot(angin)
plt.show()
```
